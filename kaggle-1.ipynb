```python
print(df.type.unique())
```

```python
Movie=df[df.type=="Movie"]
TV_Show=df[df.type=="TV_Show"]


```

```python
plt.plot(Movie.release_year,Movie.duration,color="red",label= "Movie")
plt.plot(TV_Show.release_year,TV_Show.duration,color="green",label= "TV_Show")
plt.legend()
plt.xlabel("Movie")
plt.ylabel("TV_Show")
plt.show()
```

```python
plt.scatter(Movie.release_year,Movie.duration,color="green",label= "Movie")
plt.scatter(TV_Show.release_year,TV_Show.duration,color="blue",label= "TV_Show")
plt.legend()
plt.xlabel("Movie")
plt.ylabel("TV_Show")
plt.title("scatter plot")
plt.show()
```
